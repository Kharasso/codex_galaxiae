```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Read in HYG Dataset

```python
hyg = pd.read_csv('hygdata_v3.csv')
hyg = hyg.drop(columns = ['hr', 'gl', 'bf', 'proper','bayer','flam', 'base', 'var','var_min','var_max'])
hyg = hyg.dropna()

hyg.head(10)
```

### Choose every 100th, inspect the quantiles of distance, create a small dataset

```python
print(hyg[['dist']].max())
print(hyg[['dist']].quantile(0.92))
```

```python
hyg_small = hyg[hyg.dist <= hyg.dist.quantile(0.92)].iloc[::100, :]
print(hyg_small.columns)
hyg_small.head(10)
```

```python
hyg_small.to_csv('hyg_exp.csv')
```
